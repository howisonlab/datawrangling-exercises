```python
import psycopg2

conn = psycopg2.connect(host='localhost')

# will raise ActiveSqlTransaction exception otherwise
conn.set_isolation_level( extensions.ISOLATION_LEVEL_AUTOCOMMIT )

cursor = conn.cursor()

cursor.execute("CREATE DATABASE class_objects")

conn.close()
```

```python
with psycopg2.connect(host='localhost', dbname="class_objects") as conn:
    with conn.cursor() as cursor:
        with open('create_objects.sql', 'r') as sqlfile:
            cursor.execute(sqlfile.read())
```
