```mysql
CREATE DATABASE objects_db;
```

```mysql
DROP TABLE IF EXISTS objects;
DROP TABLE IF EXISTS colors;
DROP TABLE IF EXISTS functions;
```

```mysql
CREATE TABLE objects (
    id SERIAL PRIMARY KEY,
    name TEXT,
    color_id INTEGER,
    function_id INTEGER
);
    
```

```mysql
CREATE TABLE colors (
    id SERIAL PRIMARY KEY,
    name TEXT
);
```

```mysql
CREATE TABLE functions (
    id SERIAL PRIMARY KEY,
    name TEXT
);
```
